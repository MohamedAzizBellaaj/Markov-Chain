```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# We will define phi(x)


def phi(x, a):
    return (1 / 2) * (x + a / x)
```

#### III) Write a code that generate the nth term of the sequence above for a given initial value x0 and a given positive number a > 0.


```python
def get_x_nth(x0, n, a):
    return x0 if (n == 0) else get_x_nth(phi(x0, a), n - 1, a)
```

#### IV) Plot the dynamics of your recursive sequence, and compare it with the value of √a.


```python
# We can take a = 2

a = 4
n = np.arange(1, 11, 1)
y = []

y.append(n[0])

for i in range(1, len(n)):
    y.append(phi(y[i - 1], a))

plt.axhline(y=np.sqrt(a), color="red", alpha=0.5, linestyle="-", label="√a")

plt.plot(n, y, alpha=0.5)

plt.xlabel("N")
plt.ylabel("Y-axis")
plt.title("Showing the convergence towards √a")

plt.legend()

plt.show()
```
